```python
import lxml.etree as etree
import re
from io import BytesIO
replace_xmlns = lambda s : re.sub(r'xmlns:(?P<label>\w*)\=\"(?P<val>.*)\"', r'xmlns:\g<label>="\g<label>"', s)
test_xmlns = '\txmlns:excerpt="http://wordpress.org/export/1.2/excerpt/"'
replace_xmlns(test_xmlns)
```

```python
replace_encoding = lambda s : re.sub(".*encoding\=.*", "", s)
xml_string = open('greenkeyintranet.xml').read()
pre_chan, chan, post_chan = xml_string.partition("<channel>")
pre_chan = replace_encoding(pre_chan)
pre_chan = replace_xmlns(pre_chan)
pre_chan
```

```python
xml_string = pre_chan + chan + post_chan
root = etree.XML(xml_string)
```

```python
root.nsmap = {k:k for k,v in root.nsmap.items()}
root.nsmap
```

```python
chan = root[0]
len(chan)
```

```python
items = [e for e in chan.getchildren() if e.tag=='item']
```

```python
item_dict = lambda item : {e.tag:e.text for e in item}
```

```python
[item_dict(item) for item in items[:10]]
```
